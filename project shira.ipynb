```python
import pandas as pd
import zipfile

# Path to the ZIP file
zip_file_path = r"C:\Users\user\Documents\nituach netunim\archive.zip"

# Extract the CSV file from the ZIP archive and read it into a pandas DataFrame
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # List all files in the ZIP archive
    print(z.namelist())
    
    # Assuming there's only one CSV file in the archive or we know its name
    csv_file_name = z.namelist()[0]
    
    # Read the CSV file into a pandas DataFrame
    with z.open(csv_file_name) as f:
        df = pd.read_csv(f)

# Display basic information about the DataFrame
print(df.info())

# Display basic statistics about the DataFrame
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Display the first few rows of the DataFrame
df.head()
```

##3. השפעת מאפייני השירות על שביעות הרצון

#ננתח את הקשר בין מאפייני השירות (כגון זמן אספקה, איכות האריזה, רמת הנימוס של השליח) לשביעות הרצון של הלקוחות.

```python
# Convert relevant columns to numeric scales for analysis
importance_mapping = {
    'Unimportant': 1,
    'Slightly Important': 2,
    'Moderately Important': 3,
    'Important': 4,
    'Very Important': 5
}

df['Less Delivery time'] = df['Less Delivery time'].map(importance_mapping)
df['High Quality of package'] = df['High Quality of package'].map(importance_mapping)
df['Politeness'] = df['Politeness'].map(importance_mapping)

# Plot the relationship between service attributes and Output
sns.boxplot(x='Output', y='Less Delivery time', data=df)
plt.title('Relationship between Less Delivery Time and Customer Satisfaction')
plt.show()

sns.boxplot(x='Output', y='High Quality of package', data=df)
plt.title('Relationship between High Quality of Package and Customer Satisfaction')
plt.show()

sns.boxplot(x='Output', y='Politeness', data=df)
plt.title('Relationship between Politeness and Customer Satisfaction')
plt.show()
```
