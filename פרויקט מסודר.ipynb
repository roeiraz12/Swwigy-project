נציג גם מפה כללית של חווית לקוח

```python
#השוואת חווית לקוח כללית לעמודות אחרות באמצעות heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# קריאת הנתונים
df = pd.read_csv('onlinedeliverydata.csv')

# המרת ערכים טקסטואליים לערכים מספריים בעמודת 'Output'
df['Output'] = df['Output'].astype('category').cat.codes

# רשימת העמודות להשוואה
columns_to_compare = [
    'Ease and convenient', 'Time saving', 'More restaurant choices', 
    'Easy Payment option', 'More Offers and Discount', 'Good Food quality', 
    'Good Tracking system', 'Late Delivery', 'Poor Hygiene', 
    'Bad past experience', 'Unaffordable', 'Wrong order delivered', 
    'Missing item', 'Influence of rating', 'Politeness'
]

# המרת עמודות להשוואה לערכים מספריים
for col in columns_to_compare:
    df[col] = df[col].astype('category').cat.codes

# חישוב מטריצת המתאם
correlation_matrix = df[columns_to_compare + ['Output']].corr()

# תצוגת heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Customer Satisfaction Factors')
plt.show()
```

נציג גרף רדאר של חווית לקוח כללית:

```python
import numpy as np
import matplotlib.pyplot as plt

# נתוני השביעות רצון
categories = ['Ease and convenient', 'Time saving', 'More restaurant choices', 'Easy Payment option', 
              'More Offers and Discount', 'Good Food quality', 'Good Tracking system', 'Self Cooking', 
              'Health Concern', 'Late Delivery', 'Poor Hygiene', 'Bad past experience', 'Unavailability', 
              'Unaffordable', 'Long delivery time', 'Delay of delivery person getting assigned', 
              'Delay of delivery person picking up food', 'Wrong order delivered', 'Missing item', 
              'Order placed by mistake', 'Influence of time', 'Residence in busy location', 
              'Google Maps Accuracy', 'Good Road Condition', 'Low quantity low time', 
              'Delivery person ability', 'Influence of rating', 'Less Delivery time', 
              'High Quality of package', 'Number of calls', 'Politeness', 'Freshness ', 'Temperature', 
              'Good Taste ', 'Good Quantity']

# נתוני השביעות רצון - דוגמה, אפשר לשנות את הערכים על פי הנתונים הממוצעים שברשותך
values = [4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5]

# נתוני השביעות רצון - ערכי הזוויות במעלות
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()

# סגנון הגרף
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
plt.title('Radar Chart for Overall Satisfaction', size=20, color='blue', y=1.1)

# ציור הקווים של הרדאר
ax.plot(angles, values, linewidth=1, linestyle='solid', color='blue')

# ציור האזור מתחת לקווים
ax.fill(angles, values, 'blue', alpha=0.1)

# סימון של כל קטגוריה
ax.set_yticklabels([])
ax.set_xticks(angles)
ax.set_xticklabels(categories, fontsize=10)

# הצגת הגרף
plt.show()
```

מפת חום של שביעות רצון לפי אזורים גיאוגרפים:

```python
#יצירת מפת חום באמצעות Folium
import pandas as pd
import folium
from folium.plugins import HeatMap

# קריאת הנתונים
df = pd.read_csv('onlinedeliverydata.csv')

# הכנה של הנתונים - הכפלת רשומות לפי מספר ההזמנות
df['Number of calls'] = pd.to_numeric(df['Number of calls'], errors='coerce').fillna(1)
data = df[['latitude', 'longitude', 'Number of calls']]

# יצירת רשימה של רשומות עם קואורדינטות משוכפלות לפי מספר ההזמנות
heat_data = []

for index, row in data.iterrows():
    for _ in range(int(row['Number of calls'])):
        heat_data.append([row['latitude'], row['longitude']])

# יצירת מפת בסיס
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=12)

# הוספת מפת חום
HeatMap(heat_data).add_to(m)

# שמירת המפה כקובץ HTML
m.save('heatmap.html')

# להצגה ישירה בתוך Jupyter
m
```

משהו נוסף:

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# קריאת הנתונים
df = pd.read_csv('onlinedeliverydata.csv')

# המרת עמודת Output לערכים מספריים
output_mapping = {
    'Very Satisfied': 5,
    'Satisfied': 4,
    'Neutral': 3,
    'Dissatisfied': 2,
    'Very Dissatisfied': 1
}
df['Output'] = df['Output'].replace(output_mapping)

# יצירת אפליקציית Dash
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Customer Satisfaction Dashboard"),
    dcc.Dropdown(
        id='category-dropdown',
        options=[{'label': cat, 'value': cat} for cat in df.columns if cat != 'Output'],
        value='Ease and convenient'
    ),
    dcc.Graph(id='satisfaction-graph')
])

@app.callback(
    Output('satisfaction-graph', 'figure'),
    [Input('category-dropdown', 'value')]
)
def update_graph(selected_category):
    fig = px.scatter(df, x=selected_category, y='Output', trendline='ols', title=f'Satisfaction vs {selected_category}')
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
