```python
import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency

# Assuming 'df' is your DataFrame containing the data

# Calculate expected frequencies for Meal(P1) by Age
ct_P1 = pd.crosstab(df['Perference(P1)'], df['Age'])
ct_P1_expected = pd.crosstab(df['Perference(P1)'], df['Age'], normalize='index').mul(ct_P1.sum(axis=1), axis=0)

# Plot heatmap for observed frequencies of Meal(P1) by Age
sns.heatmap(ct_P1, annot=True, cmap='YlGnBu', cbar=False)
plt.title('Observed Frequencies of Preferred Meal(P1) by Age')
plt.show()

# Plot heatmap for expected frequencies of Meal(P1) by Age
sns.heatmap(ct_P1_expected, annot=True, cmap='YlGnBu', cbar=False)
plt.title('Expected Frequencies of Preferred Meal(P1) by Age')
plt.show()

# Perform Chi-square test for Meal(P1) by Age
chi2, p, dof, expected = chi2_contingency(ct_P1)
print(f"Chi-square test statistic: {chi2}")
print(f"P-value: {p}")

# Calculate expected frequencies for Meal(P2) by Age
ct_P2 = pd.crosstab(df['Perference(P2)'], df['Age'])
ct_P2_expected = pd.crosstab(df['Perference(P2)'], df['Age'], normalize='index').mul(ct_P2.sum(axis=1), axis=0)

# Plot heatmap for observed frequencies of Meal(P2) by Age
sns.heatmap(ct_P2, annot=True, cmap='YlGnBu', cbar=False)
plt.title('Observed Frequencies of Preferred Meal(P2) by Age')
plt.show()

# Plot heatmap for expected frequencies of Meal(P2) by Age
sns.heatmap(ct_P2_expected, annot=True, cmap='YlGnBu', cbar=False)
plt.title('Expected Frequencies of Preferred Meal(P2) by Age')
plt.show()

# Perform Chi-square test for Meal(P2) by Age
chi2, p, dof, expected = chi2_contingency(ct_P2)
print(f"Chi-square test statistic: {chi2}")
print(f"P-value: {p}")


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# קריאת הנתונים
df = pd.read_csv(r"C:\Users\user\Downloads\archive (1).zip")

# המרת העמודות לערכים בולטים
replacement_dict = {
    'Agree': 'Yes',
    'Disagree': 'No',
    'Neutral': 'No',
    'Strongly agree': 'Yes',
    'Strongly disagree': 'No'
}
issues_columns = ['Late Delivery', 'Poor Hygiene', 'Bad past experience', 'Unavailability', 'Unaffordable', 'Long delivery time', 'Wrong order delivered', 'Missing item']
df[issues_columns] = df[issues_columns].replace(replacement_dict)

# גרף של תרשים עמודות
plt.figure(figsize=(12, 8))
df_age_grouped = df.groupby('Age')[issues_columns].apply(lambda x: x.apply(pd.Series.value_counts)).unstack()
df_age_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Percentage Distribution of Issues by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Issue')
plt.show()

```
